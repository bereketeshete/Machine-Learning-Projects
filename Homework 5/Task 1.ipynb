```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("MNIST_train.csv") #load the data into a pandas dataframe
data.head(5) #Show the first 5 rows
```

```python
pca_model = PCA(n_components = 50)
```

```python
pca_reduced = pca_model.fit_transform(data)
```

```python
print(data.shape)
print(pca_reduced.shape)
```

```python
kmeans_model = KMeans(n_clusters = 10)
label = kmeans_model.fit_predict(pca_reduced)
```

```python
#Getting unique labels
 
u_labels = np.unique(label)
 
#plotting the results:
 
for i in u_labels:
    plt.scatter(pca_reduced[label == i , 0] , pca_reduced[label == i , 1] , label = i, edgecolors='white')
plt.title('K-means on PCA Reduced MNIST Dataset to 50 featues with 10 Clusters')
plt.legend()
plt.show()
```
